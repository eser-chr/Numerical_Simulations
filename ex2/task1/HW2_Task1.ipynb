```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('benchmark_results.csv')

# Plot the data
plt.plot(data['Vector Size'], data['FLOPS/SEC.'], marker='o', linestyle='-', color='b')
plt.xlabel('Vector Size')
plt.ylabel('FLOPS/SEC.')
plt.title('Vector Triad Benchmark Results')
plt.grid(True)

# Add vertical lines for cache sizes if needed
# Example: plt.axvline(x=cache_size, color='r', linestyle='--', label='Cache Size')

#Show the plot in the Jupyter Notebook
plt.show()
```
